```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import LocalOutlierFactor
import numpy as np
from scipy.stats import zscore
```

```python
data = pd.read_csv('engine_data.csv')
```

```python
data.head()
```

```python
data['Temperature_difference'] = data['Coolant temp'] - data['lub oil temp']
```

```python
z_scores = np.abs(zscore(data.drop(columns=['Engine Condition'])))  # Exclude the target column for outlier detection

# Set a z-score threshold (commonly 3) to define outliers
threshold = 3
outliers = (z_scores > threshold)

# Remove rows with any outliers
data_no_outliers = data[(~outliers).all(axis=1)]

print("Original dataset size:", data.shape[0])
print("Dataset size after removing outliers:", data_no_outliers.shape[0])
print("\nData with outliers removed:")
print(data_no_outliers)
```

```python
output_path = '/content/cleaned_engine_data.csv'  # Specify a file path within the directory
data_no_outliers.to_csv(output_path, index=False)

print(f"Cleaned dataset saved to {output_path}")
```
